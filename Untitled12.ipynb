```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt # importa las bibliotecas
# calcula el polinomio de Hermite usando los datos proporcionados
def hermite_interpolation(x_data, y_data, dy_data):
  # lista de los puntos x y y, lista de las derivadas en cada punto x
    n = len(x_data)
    x = sp.symbols('x') # la x con la que vamos a construir el polinomio

    # listas para almacenar los términos del polinomio
    terms = []

    # calcula los coeficientes para el polinomio
    for i in range(n):
        # inicializa el término con el valor de la función
        term = y_data[i]
        # factor base para el punto x_i
        H = 1

        for j in range(n): # calcula el producto de (x - x_j) / (x_i - x_j) para todos j ≠ i
            if j != i:
                H *= (x - x_data[j]) / (x_data[i] - x_data[j])

        # derivada con respecto a x
        H_prime = H.diff(x)

        # añade el término de la derivada multiplicado por f'(x_i)
        term += dy_data[i] * H_prime
        terms.append(term) # almacenamos el término

    # sumamos todos los términos para obtener el polinomio
    hermite_poly = sum(terms)
    return hermite_poly

def main():
    # solicita los datos al usuario
    n = int(input("¿Cuántos puntos de datos deseas ingresar? "))

    # lstas para almacenar los datos de x, y, y las derivadas de y
    x_data = []
    y_data = []
    dy_data = []

    for i in range(n):
        x_i = float(input(f"Ingrese el valor de x_{i+1}: "))
        y_i = float(input(f"Ingrese el valor de f(x_{i+1}) = y_{i+1}: "))
        dy_i = float(input(f"Ingrese el valor de la derivada f'(x_{i+1}): "))

        x_data.append(x_i)
        y_data.append(y_i)
        dy_data.append(dy_i)

    # calcula el polinomio de Hermite
    hermite_poly = hermite_interpolation(x_data, y_data, dy_data)

    # muestra el polinomio
    print(f"El polinomio de Hermite es: {hermite_poly}")

    # grafica el polinomio junto con los datos de la tabla
    x_vals = np.linspace(min(x_data) - 1, max(x_data) + 1, 400)
    # se evalúa el polinomio en cada valor de x
    y_vals = np.array([float(hermite_poly.subs('x', val).evalf()) for val in x_vals])

    # grafica la función
    plt.plot(x_vals, y_vals, label="Polinomio de Hermite", color="orange")

    # grafica los puntos de datos originales
    plt.scatter(x_data, y_data, color="pink", zorder=5, label="Puntos de datos")

    # resalta las derivadas usando líneas de tangentes
    for i in range(n):
        tangent_x_vals = np.linspace(x_data[i] - 0.5, x_data[i] + 0.5, 10)
        tangent_y_vals = [float(hermite_poly.subs('x', x_val).evalf()) for x_val in tangent_x_vals]
        plt.plot(tangent_x_vals, tangent_y_vals, 'r--', alpha=0.7)

    # muestra la leyenda
    plt.legend()

    # etiquetas
    plt.title("Polinomio de Hermite y puntos de datos")
    plt.xlabel("x")
    plt.ylabel("f(x)")

    # muestra la gráfica
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```
