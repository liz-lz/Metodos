```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt # importa las bibliotecas

def newton_interpolation(x_values, y_values): # para calcular el polinomio dado un conjunto de puntos igualmente espaciados
   # lista de valores de x y y
    step = x_values[1] - x_values[0] # verifica que los datos estén igualmente espaciados
    if not all(sp.Abs(x_values[i+1] - x_values[i] - step) < 1e-9 for i in range(len(x_values) - 1)):
        raise ValueError("Los datos no están igualmente espaciados.")

    n = len(x_values)
    x = sp.Symbol('x')

    # crea la tabla de diferencias divididas calculadas para cada nivel
    diff_table = [y_values[:]]  # primera columna

    for j in range(1, n):
        col = []
        for i in range(n - j):
            diff = (diff_table[j-1][i+1] - diff_table[j-1][i]) / step
            col.append(diff)
        diff_table.append(col)  # matriz que almacena las diferencias divididas
        # agrega la columna
        # cada columna se calcula restando valores consecutivos de la columna anterior y dividiéndolos

    # construye el polinomio interpolante
    polynomial = diff_table[0][0] # primer término
    product_terms = 1
    # el polinomio se construye sumando los términos
    for j in range(1, n):
        product_terms *= (x - x_values[j-1]) # producto acumulativo
        polynomial += diff_table[j][0] * product_terms #suma de término con ceoficientes

    return sp.simplify(polynomial)  # simplifica el polinomio

def plot_interpolation(x_values, y_values, polynomial):
    # grafica el polinomio y los puntos originales

    x = sp.Symbol('x')
    f_lambdified = sp.lambdify(x, polynomial, 'numpy')

    # genera puntos para la gráfica del polinomio
    x_plot = np.linspace(min(x_values), max(x_values), 500)
    y_plot = f_lambdified(x_plot)

    # para la gráfica
    plt.figure(figsize=(8, 6))
    plt.plot(x_plot, y_plot, label="Polinomio de interpolación de Newton", color="purple")
    plt.scatter(x_values, y_values, color="pink", label="Puntos originales", zorder=5)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Interpolación de Newton")
    plt.legend()
    plt.grid()
    plt.show()

# prueba del código
if __name__ == "__main__":
    # solicita los datos al usuario
    print("Ingrese los valores de x separados por comas:")
    x_values = list(map(float, input().split(',')))

    print("Ingrese los valores de y separados por comas:")
    y_values = list(map(float, input().split(',')))

    if len(x_values) != len(y_values):
        print("El número de valores de x y y debe coincidir.")
    else:
        try:
            polynomial = newton_interpolation(x_values, y_values)
            print("El polinomio de Newton es:")
            print(polynomial)

            # grafica
            plot_interpolation(x_values, y_values, polynomial)
        except ValueError as e:
            print(e)

```
