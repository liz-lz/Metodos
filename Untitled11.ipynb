```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt   #importar las bibliotecas

# función para calcular el polinomio de interpolación de Lagrange
def lagrange_interpolation(x_vals, y_vals):
    x = sp.symbols('x') # la variable
    n = len(x_vals) # puntos de entrada
    L = 0  # inicia el polinomio en 0

    # calculamos el polinomio
    for i in range(n):
        # producto de los términos (x - xj)/(xi - xj)
        term = y_vals[i] #el valor de y en el punto (x_i, y_i)
        for j in range(n):
            if j != i: # no se toma el punto i cuando j = i
                term *= (x - x_vals[j]) / (x_vals[i] - x_vals[j])
        L += term  # sumamos el término al polinomio total

    return L

# grafica el polinomio y los puntos de la tabla
def plot_lagrange(x_vals, y_vals, L): # lista de coordenadas de los puntos de la tabla y el polinomio
    # crea una figura para el gráfico
    plt.figure(figsize=(8, 6))

    # graficar el polinomio
    x_vals_dense = np.linspace(min(x_vals)-1, max(x_vals)+1, 500)  # puntos para graficar el polinomio de manera suave
    L_func = sp.lambdify('x', L, 'numpy')  # Convertir el polinomio simbólico a función de NumPy
    plt.plot(x_vals_dense, L_func(x_vals_dense), label='Polinomio de Lagrange', color='pink')

    # grafica los puntos de la tabla
    plt.scatter(x_vals, y_vals, color='red', label='Datos originales', zorder=5)

    # etiquetas
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.title('Polinomio de interpolación de Lagrange')

    # mostrar la gráfica
    plt.grid(True)
    plt.show()

# pedir datos al usuario
def pedir_datos():
    n = int(input("¿Cuántos puntos de datos desea ingresar? "))
    # pide al usuario ingresar cuántos puntos de datos quiere y luego, para cada punto, pide el valor de x y y
    x_vals = []
    y_vals = []

    for i in range(n): # pide al usuario que ingrese los valores para cada punto
        x = float(input(f"Ingrese el valor de x_{i+1}: "))
        y = float(input(f"Ingrese el valor de y_{i+1}: "))
        x_vals.append(x)
        y_vals.append(y)
    # almacena y devuelve los valores
    return x_vals, y_vals

# función principal
def main():
    # pedir los datos al usuario
    x_vals, y_vals = pedir_datos()

    # calcula el polinomio
    L = lagrange_interpolation(x_vals, y_vals)

    # muestra el polinomio
    print("\nPolinomio de Lagrange:")
    sp.pprint(L)

    # grafica el polinomio y los puntos
    plot_lagrange(x_vals, y_vals, L)

# ejecuta el programa
if __name__ == "__main__":
    main()
```
